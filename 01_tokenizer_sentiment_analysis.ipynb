# 1.tokenizer,构造输入

- tokenizer, model :相匹配，tokenizer outputs  => model input
- Auto*Tokenizer,Auto*Model: Generic type,Auto开头的会自适应的找到对应的model和Tokenizer,服务于model input
    - len(input_ids) == len(attention_mask)
    - tokenizer(test_senteces[0],) ==> tokenizer.\_\_call\_\_ ==>encode
    - tokenizer.encode 约等于 tokenizer.tokenize次元 + tokenizer.convert_tokens_to_ids id；缺少起止标识符101，102
    - tokenizer.decode 
    - tokenizer 的工作原理主要依赖于tokenizer.vocab字典存储了tokens ==> ids 的映射关系
        - tokenizer.special_tokens_map 记录了一些特殊符号 ==> tokens
    - attenmask = 1 的部分为句子原始部分，attemask = 0 的部分指的时padding部分 
    - truncation 与 max_length 配合一起作用

```python
test_senteces = ['today is not that bad', 'today is so bad']
model_name = 'distilbert-base-uncased-finetuned-sst-2-english'
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
```

```python
import torch
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)
```

```python
batch_input = tokenizer(test_senteces, truncation = True,  padding=True,return_tensors='pt')
```

```python
batch_input
```

```python
tokenizer.tokenize(test_senteces[0])
```

```python
tokenizer.convert_tokens_to_ids(tokenizer.tokenize(test_senteces[0]))
```

```python
tokenizer.encode(test_senteces[0])
```

```python
tokenizer(test_senteces[0],)
```

```python
tokenizer.decode([101, 2651, 2003, 2025, 2008, 2919, 102])
```

```python
tokenizer.vocab#token:id
#尾盘匹配就是unkonwn
```

```python
tokenizer.vocab_size
```

```python
tokenizer.special_tokens_map
```

```python
tokenizer.convert_tokens_to_ids(x for x in tokenizer.special_tokens_map.values())
```

```python
tokenizer(test_senteces, truncation=True, padding=True, return_tensors='pt')#指定最大长度，超长截断，补齐, padding = true 于max_length不兼容，当指定padding = True时，遍历所有句子，以所有句子的最长长度为padding的尺寸
```

```python
tokenizer(test_senteces, max_length=32, truncation=True,
          padding='max_length', return_tensors='pt')  # 指定最大长度，超长截断，补齐
```

# 2.mode，调用模型 & parse output ，输出解析

```python
import torch.nn.functional as F
```

```python
model.config
```

```python
with torch.no_grad():
    outputs = model(**batch_input)
    print(outputs)
    score = F.softmax(outputs.logits, dim=1)
    print(score)
    labels = torch.argmax(score,dim=1)
    print(labels)
    labels = [model.config.id2label[id] for id in labels.tolist()]
    print(labels)
```
